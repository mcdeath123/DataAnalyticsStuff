```python
#Youtube Global Statistics 2023
import pandas as pd

# Load the dataset
data = pd.read_csv('D:\Data Analyst portfolio\Data-Analyst-Portfolio\DataAnalyticsStuff\Global YouTube Statistics.csv',encoding='ISO-8859-1')

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
#Distribution of subscribers by category
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set_style("whitegrid")

# 1. Distribution of subscribers by category
category_subscribers = data.groupby('category')['subscribers'].sum().sort_values(ascending=False)

plt.figure(figsize=(14, 7))
category_subscribers.plot(kind='bar', color='c')
plt.title('Distribution of Subscribers by Category')
plt.ylabel('Total Subscribers (in billions)')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Here are some insights in the data viz distribution of subscribers by category:

    The "Music" category stands out as the most popular, boasting the highest number of subscribers.

    This is followed by "Entertainment", "Film & Animation", and others.

    Some categories like "Nonprofits & Activism" and "Shows" have relatively fewer subscribers.


```python
# Displaying the top 10 channels based on their video views
# Top channels by views
top_channels_by_views = data[['Youtuber', 'video views']].sort_values(by='video views', ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(x='video views', y='Youtuber', data=top_channels_by_views, palette='viridis')
plt.title('Top 10 Channels by Video Views')
plt.xlabel('Total Video Views (in hundreds of billions)')
plt.ylabel('Channel Name')
plt.tight_layout()
plt.show()
```

The visualization above are the top 10 channels based on their video views.
Here are some insights:

    "T-Series" leads with the highest number of video views.

    "Cocomelon - Nursery Rhymes" and "SET India" follow closely behind.

    It's evident that a few channels dominate in terms of video views.

```python
# 3. Correlation between subscribers and video views
correlation = data['subscribers'].corr(data['video views'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='subscribers', y='video views', data=data, color='blue', alpha=0.6)
plt.title(f'Correlation between Subscribers and Video Views (Correlation Coefficient: {correlation:.2f})')
plt.xlabel('Subscribers (in hundreds of millions)')
plt.ylabel('Video Views (in hundreds of billions)')
plt.tight_layout()
plt.show()
```

The scatter plot showcases the relationship between subscribers and video views:

    There's a strong positive correlation between the number of subscribers and video views, with a correlation coefficient of approximately 0.79. This implies that channels with more subscribers tend to have more video views, which is expected.

```python
# 4. Analysis based on the country of the Youtuber

# Total subscribers by country
country_subscribers = data.groupby('Country')['subscribers'].sum().sort_values(ascending=False)

# Total video views by country
country_views = data.groupby('Country')['video views'].sum().sort_values(ascending=False)

# Plotting the data
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))

country_subscribers.plot(kind='bar', ax=ax1, color='magenta')
ax1.set_title('Total Subscribers by Country')
ax1.set_ylabel('Subscribers (in billions)')
ax1.set_xlabel('Country')
ax1.tick_params(axis='x', rotation=45)

country_views.plot(kind='bar', ax=ax2, color='orange')
ax2.set_title('Total Video Views by Country')
ax2.set_ylabel('Video Views (in trillions)')
ax2.set_xlabel('Country')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Here's the analysis based on the country of the YouTuber:

Total Subscribers by Country:

    The United States dominates with the highest number of subscribers, followed by India.
    Brazil, the United Kingdom, and Canada also have significant subscriber counts.

Total Video Views by Country:

    Once again, the United States leads with the most video views.
    India and Brazil are also prominent in terms of total views, with other countries like the United Kingdom and Thailand following.

These visualizations offer insights into the global distribution of YouTube popularity by country. The dominance of the United States and India in both metrics is evident, showcasing the vast audience and content creation capabilities in these nations.

```python
# 1. Channel Growth - Most rapidly growing channels based on increase in subscribers for the last 30 days
top_growing_channels = data[['Youtuber', 'subscribers_for_last_30_days']].sort_values(by='subscribers_for_last_30_days', ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(x='subscribers_for_last_30_days', y='Youtuber', data=top_growing_channels, palette='plasma')
plt.title('Top 10 Rapidly Growing Channels (Last 30 days)')
plt.xlabel('Increase in Subscribers')
plt.ylabel('Channel Name')
plt.tight_layout()
plt.show()
```

Here are some insights of top 10 rapidly growing channels based on the increase in subscribers over the last 30 days:

    "MrBeast" leads the pack, showcasing significant growth in subscriber count in the past month.

    Channels like "T-Series", "Like Nastya", and "Vlad and Niki" are also experiencing rapid growth.

This provides insights into which channels are currently trending and gaining popularity at a fast pace.

```python
# 2. Channel Longevity - Distribution of channels based on their creation year
channel_creation_distribution = data['created_year'].value_counts().sort_index()

plt.figure(figsize=(14, 7))
channel_creation_distribution.plot(kind='bar', color='pink')
plt.title('Distribution of Channels by Creation Year')
plt.ylabel('Number of Channels')
plt.xlabel('Creation Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Here are some insights in the bar graph distribution of channels based on their creation year:

    A significant number of the top channels were created around 2006-2007, which corresponds to the early years of YouTube's growth.
    
    There's another noticeable peak around 2013-2014.
    
    The number of top channels created in recent years (post-2018) seems to have declined, possibly indicating the competitive nature of the platform and the dominance of already-established channels.

```python
# 3. Country Analysis - Correlation with "Gross tertiary education enrollment (%)" and "Unemployment rate"

# Correlation with Gross tertiary education enrollment
education_correlation = data['Gross tertiary education enrollment (%)'].corr(data['subscribers'])

# Correlation with Unemployment rate
unemployment_correlation = data['Unemployment rate'].corr(data['subscribers'])

education_correlation, unemployment_correlation
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Convert the dataframe to a GeoDataFrame
geometry = [Point(xy) for xy in zip(data['Longitude'], data['Latitude'])]
geo_data = gpd.GeoDataFrame(data, geometry=geometry)

# Plot the channels on a map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightgray')
geo_data.plot(ax=ax, markersize=50, color='red', alpha=0.5)
ax.set_title('Geographical Distribution of Top YouTube Channels')
plt.show()
```

Here's the geographical distribution of the top YouTube channels:

    The red dots represent the locations of the top YouTube channels based on their provided latitude and longitude.

    A concentration of top channels can be seen in North America, Europe, and parts of Asia, especially India.

This visualization gives a spatial understanding of where the majority of popular YouTube channels are based.
